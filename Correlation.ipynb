```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Creating the DataFrame
data = {
    'Market Share': [0.375, 0.399, 0.4251, 0.409, 0.44, 0.478, 0.533, 0.537, 0.547, 0.601],
    'Passengers': [23727080, 31453451, 41600088, 48955983, 60038171, 68058886, 29336170, 46606224, 76708825, 94627412],
    'Departures': [256000, 225341, 287281, 326233, 417419, 475868, 249208, 375712, 540620, 624384]
}

df = pd.DataFrame(data)

# Calculate correlation matrix
correlation_matrix = df.corr()

# Display the correlation matrix
print("Correlation matrix:")
print(correlation_matrix)

# Create a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, center=0)

# Add title
plt.title('Heatmap of Correlation between Market Share and Passengers', fontsize=15)

# Show the plot
plt.show()

# Save the heatmap as a PNG file
plt.savefig('Correlation.jpg')
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Creating the DataFrame
data = {
    'Market Share': [0.375, 0.399, 0.4251, 0.409, 0.44, 0.478, 0.533, 0.537, 0.547, 0.601],
    'Departures': [256000, 225341, 287281, 326233, 417419, 475868, 249208, 375712, 540620, 624384]
}

df = pd.DataFrame(data)

# Calculate correlation matrix
correlation_matrix = df.corr()

# Display the correlation matrix
print("Correlation matrix:")
print(correlation_matrix)

# Create a heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5, center=0)

# Add title
plt.title('Heatmap of Correlation between Market Share and Departures', fontsize=15)

# Show the plot
plt.show()

results_path = 'results1.png'
#print(results_path)
plt.savefig(results_path)
```
